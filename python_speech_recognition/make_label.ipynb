```python
import os

def phone_to_int(label_str, label_int, phone_list, insert_sil=False):
    """
    음소 목록을 사용하는 라벨 파일
    음소를 수치로 변환
    label_str: 문자로 작성된 레이블 파일
    label_int: 문자를 숫자로 변환한 쓰기 대상의
                라벨 파일
    phone_list: 음소 목록
    insert_sil : True이면 텍스트의 시작과 끝에
                공백 삽입
    """
    # 각 파일 열기
    with open(label_str, mode='r') as f_in, open(label_int, mode='w') as f_out:
        # 라벨 파일을 한 줄씩로드
        for line in f_in:
            # 읽은 행을 공백으로 구분,
            # 리스트형의 변수로 한다
            text = line.split()
            # 목록의 0 번째 요소는 발화 ID이므로,
            # 그대로 출력
            f_out.write('%s' % text[0])
            
            if insert_sil:
                f_out.write(' 0')
            
            # 목록의 첫 번째 요소는 문자이므로,
            # 1자씩 숫자로 바꾸기
            for u in text[1:]:
                # 음소 목록에 없으면 오류
                if not u in phone_list:
                    sys.stderr.write('phone_to_int: unknown phone %s\n' % u)
                    exit(1)
                # 음소 색인 출력
                f_out.write(' %d' % (phone_list.index(u)))
                
            if insert_sil:
                f_out.write(' 0')
            #개행
            f_out.write('\n')
            
# 메인 함수            
if __name__ =='__main__':
    # 훈련 데이터의 라벨 파일 경로
    label_train_str = './data/label/train_small/text_phone'
    # 훈련 데이터 처리 결과의 출력 대상 디렉토리
    out_train_dir = './exp/data/train_small/'
    # 개발 데이터의 라벨 파일 경로
    # (개발 데이터는 GMM-HMM에는 사용되지 않지만,
    # DNN-HMM에서 사용합니다. )
    label_dev_str = './data/label/dev/text_phone'
    # 개발 데이터 처리 결과의 출력 대상 디렉토리
    out_dev_dir = './exp/data/dev'
    # 음소 목록
    phone_file = './phones.txt'
    # 포즈를 나타내는 기호
    silence_phone = 'pau'
    # True이면 문장의 시작과 끝에 포즈 삽입
    insert_sil = True
    # 음소 목록의 시작 부분에 포즈 기호를 넣습니다.
    phone_list = [silence_phone]
    # 음소리스트 파일을 열고 phone_list에 저장
    with open(phone_file, mode='r') as f:
        for line in f:
            # 공백이나 줄 바꿈을 지우고 음소 기호를 얻습니다.
            phone = line.strip()
            # 음소 목록의 끝에 추가
            phone_list.append(phone)
    # 교육 / 개발 데이터 정보를 나열합니다.        
    label_str_list = [label_train_str, label_dev_str]
    out_dir_list = [out_train_dir, out_dev_dir]
    # 훈련 / 개발 데이터에 대해 각각 처리
    for (label_str, out_dir) in zip(label_str_list, out_dir_list):
        # 출력 디렉토리가 없으면 작성
        os.makedirs(out_dir, exist_ok=True)
        # 음소와 수치의 대응리스트를 출력
        out_phone_list = os.path.join(out_dir, 'phone_list')
        
        with open(out_phone_list, 'w') as f:
            for i, phone in enumerate(phone_list):
                # 목록에 등록 된 순서
                # 그 음소에 대응하는 수치로 한다
                f.write('%s %d\n' % (phone,i))
        # 라벨의 음소 기호를 숫자로 변환하여 출력
        label_int = os.path.join(out_dir, 'text_int')
        phone_to_int(label_str, label_int, phone_list, insert_sil)
            
```
